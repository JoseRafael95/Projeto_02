ANALISANDO OS TIPOS DE TRANSAÇÕES

Neste projeto, estamos explorando os tipos de transações que apresentam os maiores valores e identificando quais desses tipos estão mais frequentemente associados a fraudes. Em nossa análise, diferenciamos entre duas categorias: Fraud, que indica uma tentativa de fraude, e FlaggedFraud, que se refere a transações que tentam movimentar quantias superiores ao limite especificado. Essa investigação é crucial para entender melhor os padrões de comportamento fraudulentos e aprimorar as medidas de prevenção.

```python
dados_pivo = pd.pivot_table(df, index = 'type', values = ['amount', 'isFraud', 'isFlaggedFraud'], aggfunc=['sum', 'std'])
paleta = sns.light_palette("red", as_cmap=True)
dados_pivo.style.background_gradient(cmap=paleta)
```

"As transações com os maiores valores movimentados concentram-se nas categorias Transfer, Cash_out e Cash_in. Observamos que todas as tentas de fraude (isFraud) ocorreram no tipo Cash_out, enquanto as transações sinalizadas como suspeitas (isFlaggedFraud) estavam exclusivamente no tipo Transfer.

Esses dados indicam que as fraudes se manifestam principalmente nos dois tipos de transações mais comuns, sugerindo que essas categorias são especialmente vulneráveis e requerem monitoramento cuidadoso."

ANALISANDO DIFERENÇAS NO VALORES TRANSFERIDOS, POSSÍVEIS TAXAS

Alguns valores movimentados 'amount' apresentam uma difereça do valor 'oldbalanceOrg - newbalanceOrig' que é o valor que realmente saiu o entrou na conta do cliente, esse valore pode identificar algum tipo de taxa.

Vamos verificar se a presença desse valor de possivel taxa tem algum tipo de relação com as fraudes.

```python
df['taxa'] = (abs(round(df['amount'])) - (abs(round(df['oldbalanceOrg'] - df['newbalanceOrig']))) ) != 0
df['taxa'] = df['taxa'].astype(int)
df
```

```python
df_group = df.groupby(['taxa','isFraud']).size().reset_index(name='count')
df_group
```

Analisando essa tabela podemos ver que nos dados onde existem taxa o numero de fraudes é menor, isso talvez não
signifique uma relação porém seria interessante deixar essa coluna de taxa para o modelo analisar e vericaremos depois
se exitem melhora ou piora na avaliação

```python
df_group = df.groupby(['taxa','isFlaggedFraud']).size().reset_index(name='count')
df_group
```

Diferente de isFraud a variavel isFlaggedFraud tem um comportamento oposto, os dados que não possuem taxas não tem tem uma movimentação suspeita.
